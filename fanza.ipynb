```python
# DMM Webサービスを使い、商品データを取得する
#
# 参考サイト
# https://affiliate.dmm.com/api/guide/
```

```python
# 以後使う固定値をここで指定する
api_id = 'xxxxxxxxxx' 
affiliate_id = 'takashi209-990'
article = 'genre'
article_id = '2007' # 「ジャンル：ぽっちゃり」のみ取得する
```

```python
# 以降で使う関数を定義する
import json
import requests

def getItemList(offset):
    url = f'https://api.dmm.com/affiliate/v3/ItemList?api_id={api_id}&affiliate_id={affiliate_id}&site=FANZA&service=digital&floor=videoa&article={article}&article_id={article_id}&hits=100&offset={offset}&sort=review&output=json'
    result = requests.get(url)
    itemList = json.loads(result.text)
    return itemList
```

```python
# 検索対象となるデータの件数を確認する
total_count = getItemList(1)['result']['total_count']
print(f'total_count: {total_count}')
```

```python
# 検索対象を取得する
import pandas as pd
from tqdm.notebook import tqdm

df = pd.DataFrame()
for offset in tqdm(range(1, total_count, 100)):
    itemList = getItemList(offset)
    if len(itemList['result']['items']) == 0:
        break

    df_itemList = pd.DataFrame()
    for item in itemList['result']['items']:
        df_item = pd.json_normalize(item)
        df_itemList = df_itemList.append(df_item)

    df = df.append(df_itemList)
```

```python
# インデックスを採番し直す
df = df.reset_index(drop=True)
```
