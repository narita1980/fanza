```python
# パッケージの読み込み
import pandas as pd
from pycaret.regression import *
```

```python
# 使用するデータの読み込み
fanza_data = pd.read_csv('fanza_genre_2007.normalize.csv')
```

```python
fanza_data.head()
```

```python
# PyCaretを起動する
exp1 = setup(fanza_data, target = 'review.average', ignore_features = None)
```

```python
# モデルの比較
compare_models()
```

```python
# モデルを作成
model = create_model('gbr')
```

```python
save_model(model ,model_name = 'model')
```

```python
# 解析結果の確認
evaluate_model(model)
```

```python
#モデルの確定
final_model = finalize_model(model)
print(final_model)
```

```python
final_model = load_model('model')
unseen_predictions = predict_model(final_model, data = fanza_data)
```

```python
unseen_predictions.to_csv("result.csv")
```
